Graphs and charts
========

```python
Line Graphs:
    * A line chart is a graph that connects a series of points by drawing line segments between them
    * these points are ordered in one of their coordinate(usually the x-coordinate) value.
    * Line charts are usually used in identifying the trends in data
```

```python
import pandas as pd

df=pd.read_excel(r"D:\Learning\Data science\DATA\PythonSpreadSheet.xls",sheet_name=1)

print(df)

print()

import matplotlib.pyplot as plt

plt.plot(df.Age,df.Salary,"r*--",linewidth=2,markersize=12)
plt.title("Line Plot-Age vs Salary",color="Green",fontsize=20)
plt.xlabel("Age",color="Red",fontsize=12)
plt.ylabel("Salary",color="Red",fontsize=12)
plt.xlim([10,50])
plt.ylim([10000,50000])
plt.grid(True)
plt.legend(["Salary"])
plt.savefig("D:\Learning\Data science\DATA\Graph\LineGraph1.png",bbox_inches="tight",pad_inches=2,transparent=True)
plt.show()
```
