# Panel or 3 dimentional

1. Items - axis 0, each item corresponds to a DataFrame contained inside
2. major_axis - axis 1, it is the index(rows) of each of the DataFrames
3. minor_axis - axis 2, it is the columns of each of the DataFrames

Syntex: pandas.Panel(data, items, major_axis, minor_axis, dtype,copy)

*Input should be only arrays

```python
import numpy
MyData=numpy.random.randn(2,3,4)
print(MyData)

import pandas
panal_Data=pandas.Panel(data=MyData)
print(panal_Data)
```

```python
import numpy
MyData=numpy.random.randn(2,3,4)
print(MyData)

import pandas
panal_Data=pandas.Panel(data=MyData,items=["A","B"],major_axis=["R1","R2","R3"],minor_axis=["C1","C2","C3","C4"])
print(panal_Data)
```
