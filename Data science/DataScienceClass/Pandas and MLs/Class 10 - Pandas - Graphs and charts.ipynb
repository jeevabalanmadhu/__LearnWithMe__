Scatter plots
=========

```python
* Scatter points show many points plotted in the carteesion plane
* Each point represent the values of two variables
* One variable is choosen in the horizontal axis and another in the vertical acis
```

```python
import pandas as pd

df=pd.read_excel(r"D:\Learning\Data science\DATA\PythonSpreadSheet.xls",sheet_name=1)

print(df)

print()

import matplotlib.pyplot as plt

plt.scatter(df.Age,df.Salary)
plt.title("Line Plot-Age vs Salary",color="Green",fontsize=20)
plt.xlabel("Age",color="Red",fontsize=12)
plt.ylabel("Salary",color="Red",fontsize=12)
plt.xlim([10,50])
plt.ylim([10000,50000])
plt.grid(True)
plt.legend(["Salary"])
plt.savefig("D:\Learning\Data science\DATA\Graph\scatterplot1.png",bbox_inches="tight",pad_inches=3,transparent=True)
plt.show()
```

```python
import pandas as pd

df=pd.read_excel(r"D:\Learning\Data science\DATA\PythonSpreadSheet.xls",sheet_name=1)

print(df)

print()

import matplotlib.pyplot as plt

plt.scatter(df.Age,df.Salary,s=100,marker="+",color="red")
plt.title("Line Plot-Age vs Salary",color="Green",fontsize=20)
plt.xlabel("Age",color="Red",fontsize=12)
plt.ylabel("Salary",color="Red",fontsize=12)
plt.xlim([10,50])
plt.ylim([10000,50000])
plt.grid(True)
plt.legend(["Salary"])
plt.savefig("D:\Learning\Data science\DATA\Graph\scatterplot2.png",bbox_inches="tight",pad_inches=2,transparent=True)
plt.show()
```

```python
Bar chart:
    1. A bar chart represents data in rectangular bars with length of the bar propotional to the value of the variable
    2. python uses the function to create bar charts
    3. Python can draw both vertical and horizantal bars in the bar chart
    4. in bar chart each of the bars can be given different colors
```

```python
import pandas as pd

df=pd.read_excel(r"D:\Learning\Data science\DATA\PythonSpreadSheet.xls",sheet_name=1)

print(df)

print()

import matplotlib.pyplot as plt

plt.bar(df.Name,df.Salary)
plt.title("Bar-Name vs Salary",color="Green",fontsize=20)
plt.xlabel("Name",color="Red",fontsize=12)
plt.ylabel("Salary",color="Red",fontsize=12)
#plt.xlim([10,50]) *for bar chart xlin is not required as its chart is cataories
plt.ylim([10000,50000])
plt.grid(True)
plt.legend(["Salary"])
plt.savefig("D:\Learning\Data science\DATA\Graph\BarChart1.png",bbox_inches="tight",pad_inches=1,transparent=True)
plt.show()
```

Pie Chart:
   =============

```python
* Python programming language has numerous libraries to create charts and graphs
* A pie chart is a representation of values as slices of a circle with different colors
* The slices are labeled and the numbers corresponding to each slice is also represented in the chart
```

```python
import pandas as pd

df=pd.read_excel(r"D:\Learning\Data science\DATA\PythonSpreadSheet.xls",sheet_name=1)

print(df)

print()

import matplotlib.pyplot as plt

plt.pie(df.Salary,radius=3,labels=df.Name)

plt.savefig("D:\Learning\Data science\DATA\Graph\PieChart1.png",bbox_inches="tight",pad_inches=2,transparent=True)
plt.show()
```
