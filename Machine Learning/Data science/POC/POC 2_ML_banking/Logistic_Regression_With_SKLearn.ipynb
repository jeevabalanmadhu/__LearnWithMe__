```python
#Getting data from excel sheet
import pandas as pd
df=pd.read_csv("banking.csv")
df.describe()
```

```python
#Training Data
x=df.loc[:,["age","duration","campaign","pdays","previous","emp_var_rate","cons_price_idx","cons_conf_idx","euribor3m","nr_employed"]]
y=df.loc[:,['y']]
```

```python
x.head()
```

```python
y.head()
```

```python
#Data Analytics or prediction using machin learning;
#sklearn = sceince kit learn

from sklearn.linear_model import LogisticRegression

model=LogisticRegression()

model.fit(x,y)
```

```python
Goodness_Score=model.score(x,y)
print("Score or R^2 value: ",Goodness_Score)
```

```python
Salary_Prediction=model.predict([[30,68,8,999,0,1.4,93.918,-42.7,4.961,5228.1]])
print(Salary_Prediction)
```
