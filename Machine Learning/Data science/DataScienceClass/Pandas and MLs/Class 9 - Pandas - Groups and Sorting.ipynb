Group and Sorting
================

```python
* Any group operation involves one of the following operations on the original object.

They are:
    1. Splitting the object
    2. Applying a function
    3. Combining results
* In many situations, we split the data into sets and we apply some functionality on each subset.
* In the apply functionality, we can perform the following operations.
    1. Aggregation - computing a summary statistic
    2. Transformation - perform some group - specific operation
    3. Filteration - discarding the data with some condition
```

```python
import pandas as pd

df=pd.read_excel(r"D:\Learning\Data science\DATA\PythonSpreadSheet.xls",sheet_name=0)

print(df)
print(df.groupby("Name"))

print()

Group=df.groupby("Name").groups
print(Group)
```
