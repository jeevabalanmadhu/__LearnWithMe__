ReIndexing
===============

```python
* Handling missing data
1. Reorder the existing data to match a new set of labels
2. Insert missing value (NA) markers in label locations where no data for the label existed.
```

```python
import pandas as pd
df1=pd.read_excel("D:\Learning\Data science\DATA\PythonSpreadSheet.xls",sheet_name=1)
print("df1\n---------------\n",df1)

df2=pd.read_excel("D:\Learning\Data science\DATA\PythonSpreadSheet.xls",sheet_name=2)
print("df2\n---------------\n",df2)

print("\nPython pandas - Reindexing: ")
df1=df1.reindex_like(df2) # override number of rows as per DF2,if df2 has less number of rows then df1, remove remaining number of rows and save in df1
print("df1-reindex\n---------------\n",df1)
```

Renaming
===========

```python
* The rename() method allows you to relabel an axis based on some mapping (a dict or series) or an arbitary function
```

```python
import pandas as pd
df1=pd.read_excel("D:\Learning\Data science\DATA\PythonSpreadSheet.xls",sheet_name=1)
print(df1)

print("\nPython pandas - Renaming")
After_Rename=df1.rename(index={1:"A",3:"C"},columns={"Name":"FirstName"}) #index indicates row and columns indicates columns
print(After_Rename)
```

Iteration:
==========

```python
1. The behaviour of basic iteration over pandas objects depends on the type.
* Series - values
* Dataframe - column labels
* panel - item labels

Functions:
    1. iteritems() #to iterate over the (key, value) pairs
    2. iterrows() #to iterate over the rows as (index, series) pairs
    3. itertuples() #iterate over the rows as namedtuples
    
```

```python
import pandas as pd
df=pd.read_excel("D:\Learning\Data science\DATA\PythonSpreadSheet.xls",sheet_name=1)
print(df)

print("\n Pandas-iteration")

#iteritems()
for key,value in df.iteritems():
    print(key,"\n",10*"_",sep="")
    print(value)
```

```python
#iterrows()

for index,series in df.iterrows():
    print(index,"\n",5*"_",sep="")
    print(series)
```

```python
#itertuples()

for rows in df.itertuples():
    print(rows)
```
